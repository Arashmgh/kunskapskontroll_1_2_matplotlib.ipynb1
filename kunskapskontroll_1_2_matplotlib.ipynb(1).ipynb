```python
# 2. Explain shortly what a figure, axes, axis, and an artist is in Matplotlib.
Figure: Entire plotting area or window.
Axes: The plot area where data is drawn.
Axis: X or Y lines in a plot.
Artist: Any element drawn on the figure (e.g., lines, text).
```

```python
# 4. Create a plot of the function y = x^2 [from -4 to 4, hint: use the np. linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(-4, 4, 100)
y = x**2

# Object-oriented approach
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Object-Oriented: y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()

# Pyplot approach
plt.plot(x, y)
plt.title("Pyplot: y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 5. Create a figure containing 2 subplots where the first is a scatter plot and the second is a bar plot. You have the data below.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot
ax1.scatter(random_data_x, random_data_y)
ax1.set_title("Scatter Plot")

# Bar plot
ax2.bar(names, values)
ax2.set_title("Bar Plot")

plt.tight_layout()
plt.show()
```
